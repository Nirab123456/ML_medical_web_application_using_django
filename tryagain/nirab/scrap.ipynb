```python
    def get_generic_medication_details(self):
        request = self.request
        selected_strength = request.GET.get('strength')
        generic_name = request.GET.get('generic_name')
        generic_name = generic_name.lower()
        dosage_form = request.GET.get('dosage_form')
        medications = Medication.objects.filter(strength=selected_strength, generic_name=generic_name, dosage_form=dosage_form)
        if medications.exists():
            medication_details = []
            for medication in medications:
                details = {
                    'name': medication.name.strip().capitalize(),
                    'dosage_form': medication.dosage_form.strip().capitalize(),
                    'generic_name': medication.generic_name.strip().capitalize(),
                    'manufacturer': medication.manufacturer.strip().capitalize(),
                    'price': str(medication.price).strip(),
                    'price_analysis': str(medication.price_analysis).strip(),
                    # add any other fields you want to include
                }
                medication_details.append(details)

            print(f'medication details: {medication_details}')            
            return JsonResponse(medication_details, safe=False)
        else:
            return JsonResponse({'error': 'Medication not found'}, status=404)
```

```python

    function showPage(pageNumber) {
        var pageItems = getPageItems(pageNumber);
        var container = '';

        for (var i = 0; i < pageItems.length; i++) {
            var pageItem = pageItems[i];

            container += `
                <div class="custom-block custom-block-topics-listing bg-white shadow-lg mb-5">
                    <div class="d-flex">
                        <div class="custom-block-topics-listing-info d-flex">
                            <div>
                                <h5 class="mb-2">${pageItem.title}</h5>
                                <p class="mb-0">${pageItem.content}</p>
                                <a href="topics-detail.html" class="btn custom-btn mt-3 mt-lg-4">Learn More</a>
                            </div>
                            <span class="badge bg-design rounded-pill ms-auto">14</span>
                        </div>
                    </div>
                </div>`;
        }

        document.getElementById("diery_details_container").innerHTML = container;
    }

    document.addEventListener("DOMContentLoaded", function() {
        var data = JSON.parse('{{ all_diery_objects_list|safe }}');
        console.log(data);

        function calculateTotalPages(items) {
            var itemsPerPage = 3;
            var totalItems = items.length;
            return Math.ceil(totalItems / itemsPerPage);
        }

        function getPageItems(pageNumber) {
            var itemsPerPage = 3;
            var startIndex = (pageNumber - 1) * itemsPerPage;
            var endIndex = startIndex + itemsPerPage;
            return data.slice(startIndex, endIndex);
        }

        function pagination() {
            var totalPages = calculateTotalPages(data);
            var currentPage = 1;

            var paginationContainer = `
                <nav aria-label="Page navigation example">
                    <ul class="pagination justify-content-center mb-0">
            `;

            for (var i = 1; i <= totalPages; i++) {
                paginationContainer += `
                    <li onclick="showPage(${i})" class="page-item page-link">${i}</li>
                `;
            }

            paginationContainer += `
                        <li class="page-item">
                            <a class="page-link" href="#" aria-label="Next">
                                <span aria-hidden="true">Next</span>
                            </a>
                        </li>
                    </ul>
                </nav>
            `;

            document.getElementById("diry_list_pagination").innerHTML = paginationContainer;
        }

        pagination();
    });

```
