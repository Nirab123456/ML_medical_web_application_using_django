```python
import pandas as pd
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForTokenClassification

tokenizer = AutoTokenizer.from_pretrained("d4data/biomedical-ner-all")
model = AutoModelForTokenClassification.from_pretrained("d4data/biomedical-ner-all")

class DRUG_CLASS_CLASSIFICATION():

    def __init__(self, main_df):
        self.pipe = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple", device=0) # pass device=0 if using gpu
        self.main_df = main_df
        drug_class_list = main_df['use_cases'].unique().tolist()
        # Remove nan values from the list
        self.all_details = [item for item in drug_class_list if not pd.isna(item)]
    
    def classify_presciption_drugs(self):
        all_results = []
        for generic_name, use_case in zip(self.main_df['generic_name'], self.main_df['use_cases']):
            # Pass generic_name and use_case to the pipeline
            results = self.pipe(f"{generic_name} {use_case}")
            print(f'result: {results}')
            for result in results:
                if result is not None:
                    if len(str(result['word'])) >= 4 and '#' not in str(result['word']):
                        all_results.append({'Entity Group': result['entity_group'], 'Word': str(result['word']), 'Score': result['score'], 'Generic Name': generic_name, 'Use Case': use_case})
        # Return an empty DataFrame if no results are found
        if not all_results:
            return pd.DataFrame(columns=['Entity Group', 'Word', 'Score', 'Generic Name', 'Use Case'])
        return pd.DataFrame(all_results)

# Create an instance of the DRUG_CLASS_CLASSIFICATION class
drug_classifier = DRUG_CLASS_CLASSIFICATION(drug_class_df)

# Classify the prescription drugs and save results in a pandas DataFrame
new_result_df = drug_classifier.classify_presciption_drugs()
```
