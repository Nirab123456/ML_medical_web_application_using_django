```python
import pandas as pd 

disease_drug_df = pd.read_csv('raw_file/Drug_clean.csv')
disease_drug_df=disease_drug_df.astype('str').astype(str).apply(lambda x: x.str.lower())
current_main_data = pd.read_csv('current_main_data(changeable)/selected_questions.csv')
```

```python
current_generic_name = current_main_data['generic_name'].unique()
new_generic_name = disease_drug_df['Drug'].unique()

common_generic_name = list(set(current_generic_name) & set(new_generic_name))
```

```python
len(common_generic_name)
```

```python
#extract the rows from new df using common_generic
common_generic_df = disease_drug_df[disease_drug_df['Drug'].isin(common_generic_name)]
common_generic_df.head()
```

```python
common_generic_df
```

```python
selected_collum = ['Drug', 'Condition', 'Satisfaction', 'Effective']
new_common_df = common_generic_df[selected_collum]
#rename columns
new_common_df = new_common_df.rename(columns={'Drug': 'generic_name', 'Condition': 'cases', 'Satisfaction': 'rating', 'Effective': 'useful_count'})
```

```python
new_common_df.to_csv('custom/common_2.csv', index=False)
```
