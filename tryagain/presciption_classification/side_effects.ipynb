```python
import pandas as pd 
df = pd.read_csv('side_effects_.7_.csv')
```

```python
df['Entity Group'].unique()
```

```python
# Rename the columns
df.rename(columns={'Entity Group': 'group', 'Word': 'indication', 'Score': 'Score', 'Sentence': 'side_effects'}, inplace=True)
```

```python
len(df['indication'].unique())
```

```python
df.drop_duplicates(subset=['group', 'indication', 'side_effects'], keep='first', inplace=True)
```

```python
df = df.applymap(lambda x: str(x).lower())
```

```python
pattern = r'^anti\s*-?\s*$'

# Use str.contains with the pattern to filter out the rows
filtered_df = df[~df['indication'].str.contains(pattern)]
```

```python
unique_value_to_filter = 'acit'

# Step 2: Create a boolean mask
mask = filtered_df['indication'] == unique_value_to_filter

# Step 3: Use the boolean mask to drop the rows from the DataFrame
filtered_df = filtered_df[~mask]
```

```python
filtered_df
```

```python
full_df= pd.read_csv('cleaned_generic_df_main.csv')
```

```python
selected_columns = ['generic_name', 'side_effects_description']
full_new_df = full_df[selected_columns]
```

```python
full_new_df.rename(columns={'side_effects_description': 'side_effects'}, inplace=True)
```

```python
merged_df = filtered_df.merge(full_new_df, on='side_effects', how='left')
```

```python
merged_df.to_csv('side_effects_with_GN(70)_.csv', index=False)
```
