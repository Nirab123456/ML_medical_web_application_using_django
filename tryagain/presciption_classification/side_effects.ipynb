```python
import pandas as pd 
df = pd.read_csv('side_effects_.7_.csv')
```

```python
df['Entity Group'].unique()
```

```python
# Rename the columns
df.rename(columns={'Entity Group': 'group', 'Word': 'indication', 'Score': 'Score', 'Sentence': 'side_effects'}, inplace=True)
```

```python
len(df['indication'].unique())
```

```python
df.drop_duplicates(subset=['group', 'indication', 'side_effects'], keep='first', inplace=True)
```

```python
df = df.applymap(lambda x: str(x).lower())
```

```python
pattern = r'^anti\s*-?\s*$'

# Use str.contains with the pattern to filter out the rows
filtered_df = df[~df['indication'].str.contains(pattern)]
```

```python
filtered_df
```

```python
filtered_df.to_csv('final_django_side_effects_.7_results.csv', index=False)
```
