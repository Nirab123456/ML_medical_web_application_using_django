```python
medicine_df = pd.read_csv('medicine.csv')
# Replace NaN values in 'Package Size' with 'package container' values
medicine_df['Package Size'].fillna(medicine_df['package container'], inplace=True)
remove_collumns = ['brand id', 'slug', 'package container']
new_medicine_df = medicine_df.drop(remove_collumns, axis=1)
df= new_medicine_df
df = df.applymap(lambda x: str(x).lower())
```

```python
# Remove parentheses from the 'Column1' column
df['Package Size'] = df['Package Size'].str.replace(r"\(|\)", "",regex=True)
# Remove everything after the first comma in 'Column1'
df['Package Size'] = df['Package Size'].str.split(',', n=1).str[0]
new_columns = {'brand name': 'name',
               'type': 'Type',
               'dosage form': 'dosage_form',
               'generic': 'generic_name',
               'strength': 'strength',
               'manufacturer': 'manufacturer',
               'Package Size': 'price'}

# Rename the columns
df = df.rename(columns=new_columns)
df.to_csv('django_medicine_data_1.csv', index=False)
```

```python
df= pd.read_csv('django_medicine_data_1.csv')
```

```python
df = df.applymap(lambda x: str(x).lower())
# Remove parentheses from the 'Column1' column
df['price'] = df['price'].str.replace(r"৳", "",regex=True)
# Remove parentheses from the 'Column1' column
df['price_analysis'] = df['price'].str.replace(r"'s pack", "",regex=True)
def calculate_price(row):
    price = row['price_analysis']
    if ':' in price:
        parts = price.split(':')
        if len(parts) == 2 and parts[0].strip().isdigit() and parts[1].strip().replace('.', '').isdigit():
            value = float(parts[1].strip()) / float(parts[0].strip())
            return f"{value:.2f}"
    return price

# Apply the function to the 'price' column
df['price_analysis'] = df.apply(calculate_price, axis=1)
# Remove everything before the colon in 'price' column
df['price_analysis'] = df['price_analysis'].apply(lambda x: x.split(':')[1].strip() if ':' in x else x)

```
