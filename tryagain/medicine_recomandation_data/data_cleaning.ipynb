```python
import pandas as pd 

df_1 = pd.read_csv('all_page_except_price_7.csv')
df_2 = pd.read_csv('all_page_except_price_8.csv')
df_3 = pd.read_csv('all_page_except_price_9.csv')
```

```python
df_1
```

```python
full_df = pd.concat([df_1, df_2, df_3])
full_df = full_df.reset_index(drop=True)
```

```python
df=full_df
```

```python
df['Product Name'] = df['link'].apply(lambda x: x.split('/')[-1].split(',')[0])
# Drop the column
df = df.drop('link', axis=1)
df = df.applymap(lambda x: str(x).lower())
df['name'] = df.apply(lambda row: row['Product Name'].rsplit(row['strength'].split()[0], 1)[0].rstrip('-'), axis=1)
df = df.drop('Product Name', axis=1)
df.to_csv('cleaned_data.csv')
len(df['dosage_form'].unique()), len(df['strength'].unique()), len(df['generic_name'].unique()), len(df['name'].unique())
df = df.drop_duplicates()
# Add a column called 'price' and set all values to zero
df['price'] = 0

```

```python
df.to_csv('cleaned_2.csv')
```
