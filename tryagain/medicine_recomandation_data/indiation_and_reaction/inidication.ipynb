```python
generic_df = pd.read_csv('generic.csv')
```

```python
new_generic_df =generic_df.replace(r'<[^>]*>', ' ', regex=True)
```

```python
new_generic_df = new_generic_df.applymap(lambda x: str(x).lower())
```

```python
unique_indection_list
```

```python
def save_wordlist_to_text(wordlist, output_file):
    try:
        with open(output_file, "w") as file:
            for word in wordlist:
                file.write(word + "\n")
        print("Wordlist saved to", output_file)
    except IOError as e:
        print("Error while saving the wordlist:", str(e))


# Save the wordlist to a text file named "wordlist.txt"
save_wordlist_to_text(unique_indection_list, "unique_indection_list.txt")
```

```python
selected_collum=['generic name','drug class','indication','indication description','therapeutic class description','pharmacology description','dosage description','interaction description','contraindications description','side effects description']

cleaned_generic_df=new_generic_df[selected_collum]
```

```python

# List of selected columns
selected_collum=['generic name','drug class',
                 'indication','indication description',
                 'therapeutic class description','pharmacology description',
                 'dosage description','interaction description',
                 'contraindications description','side effects description']

# Rename the columns and replace spaces with underscores
cleaned_generic_df.rename(columns=lambda x: x.replace(' ', '_'), inplace=True)

# Filter the DataFrame to keep only the selected columns
cleaned_generic_df = cleaned_generic_df[selected_collum]
```

```python
cleaned_generic_df.to_csv('cleaned_generic_df_main.csv',index=False)
```

```python
# Function to remove non-ASCII characters
def remove_non_ascii(text):
    return text.encode('ascii', 'ignore').decode('ascii')

# Apply the function to remove non-ASCII characters from all cells in the DataFrame
cleaned_generic_df = cleaned_generic_df.applymap(remove_non_ascii)
```

```python
cleaned_generic_df.to_csv('cleaned_generic_df_2.csv',index=False)
```
