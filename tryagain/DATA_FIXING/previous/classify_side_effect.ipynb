```python
import pandas as pd

```

```python
from transformers import pipeline
tqa = pipeline(task="question-answering",model='deepset/bert-large-uncased-whole-word-masking-squad2',device=0)
df = pd.read_csv('side_effects.csv')
side_effect_list = df['side_effects'].unique().tolist()
df['related_results'] = ""

# Loop through each unique side effect and get related results
side_effect_list = df['side_effects'].unique().tolist()
for side_effect in side_effect_list:
    context = side_effect
    question = "classify the medical terms"
    result = tqa(question=question, context=context)

    # Store the result in the 'related_results' column corresponding to the side effect
    df.loc[df['side_effects'] == side_effect, 'related_results'] = result['answer']

# Save the DataFrame with the related results to a new CSV file
df.to_csv('side_effects_with_results.csv', index=False)
```

```python
from transformers import pipeline
tqa = pipeline(task="question-answering",model='deepset/bert-large-uncased-whole-word-masking-squad2',device=0)
df = pd.read_csv('cleaned_generic_df_main.csv')

for i in range(0, len(df)):
    context = df['contraindications_description'][i]
    question = "what are the contraindications?"
    result = tqa(question=question, context=context)
    df['related_results'][i] = result['answer']
```
