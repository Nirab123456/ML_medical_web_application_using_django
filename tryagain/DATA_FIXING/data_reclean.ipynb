```python
import pandas as pd 

previous_cleaned_df = pd.read_csv('main_data/cleaned_generic_df_main.csv')
previous_cleaned_df.head()
```

```python
previous_cleaned_df.fillna('no data available', inplace=True)
```

```python
previous_cleaned_df = previous_cleaned_df.astype(str).apply(lambda x: x.str.lower())
```

```python
previous_cleaned_df
```

```python
import re
def clean_text(cell_text):
    # Remove newlines using regex
    cell_text = re.sub(r'\n', ' ', cell_text)
    
    # Remove extra spaces using regex
    cell_text = re.sub(r'\s+', ' ', cell_text)
    
    return cell_text.strip()

# Apply the clean_text function to the entire DataFrame
cleaned_df = previous_cleaned_df.applymap(clean_text)

print(cleaned_df)
```

```python
cleaned_df.to_csv('new_data/cleaned_generic_df_main_1.csv', index=False)
```
