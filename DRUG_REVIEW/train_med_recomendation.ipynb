```python
df['condition'] = df['condition'].str.replace(r"2</span> users found this comment helpful.", 'found helpful',regex=True)
```

```python
def save_wordlist_to_text(wordlist, output_file):
    try:
        with open(output_file, "w") as file:
            for word in wordlist:
                file.write(word + "\n")
        print("Wordlist saved to", output_file)
    except IOError as e:
        print("Error while saving the wordlist:", str(e))


# Save the wordlist to a text file named "wordlist.txt"
save_wordlist_to_text(unique_condition, "unique_condition.txt")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


# Create a DataFrame
labeled_df = pd.DataFrame(df)

# Step 1: Label Encoding
encoder_drugName = LabelEncoder()
encoder_condition = LabelEncoder()

labeled_df['encoded_drugName'] = encoder_drugName.fit_transform(df['drugName'])
labeled_df['encoded_condition'] = encoder_condition.fit_transform(df['condition'])

# Create dictionaries to map the encoded values back to the original values
drugName_mapping = dict(zip(encoder_drugName.classes_, encoder_drugName.transform(encoder_drugName.classes_)))
condition_mapping = dict(zip(encoder_condition.classes_, encoder_condition.transform(encoder_condition.classes_)))

# Print the DataFrame with the new encoded columns
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import json



# Create a DataFrame
labeled_df = pd.DataFrame(df)

# Step 1: Label Encoding
encoder_drugName = LabelEncoder()
encoder_condition = LabelEncoder()

labeled_df['encoded_drugName'] = encoder_drugName.fit_transform(df['drugName'])
labeled_df['encoded_condition'] = encoder_condition.fit_transform(df['condition'])

# Create dictionaries to map the encoded values back to the original values
drugName_mapping = {key: int(value) for key, value in zip(encoder_drugName.classes_, encoder_drugName.transform(encoder_drugName.classes_))}
condition_mapping = {key: int(value) for key, value in zip(encoder_condition.classes_, encoder_condition.transform(encoder_condition.classes_))}

# Save the mappings to a JSON file
with open('drugName_mapping.json', 'w') as f:
    json.dump(drugName_mapping, f)

with open('condition_mapping.json', 'w') as f:
    json.dump(condition_mapping, f)

# Print the DataFrame with the new encoded columns
print(labeled_df)
```

```python
labeled_df.to_csv('django_train_data_with_label.csv', index=False)
```

```python
# Custom function to find and replace words
def replace_found_helpful(text):
    if 'found' in text.lower() and 'helpful' in text.lower():
        return 'found helpful'
    return text

# Apply the custom function to the 'text' column
labeled_df['condition'] = labeled_df['condition'].apply(replace_found_helpful)

```
