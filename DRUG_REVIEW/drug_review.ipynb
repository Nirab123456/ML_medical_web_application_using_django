```python

# Initialize a dictionary to store the word counts
word_counts = {}

# Iterate over the search words
for word in list_of_generic:
    # Count the occurrences of the word in the 'Text' column
    counts = df['generic_name'].str.count(word).sum()
    
    # Store the count in the dictionary
    word_counts[word] = counts

# Print the word counts
```

```python
word_list=[]
for word, count in word_counts.items():
    if count > 0:
        print(f"Word: {word}, Count: {count}")
        word_list.append(word)
```

```python
len(word_list)
```

```python
filtered_df = df_1[df_1['drugName'].isin(word_list)]
new_df = filtered_df.reset_index(drop=True)

```

```python
word_list= pre_processed_df['condition'].unique()
```

```python
# Create an empty list to store the new rows
new_rows = []

# Iterate over each drug name and word in the word list
for drug in pre_processed_df['drugName']:
    for word in word_list:
        new_row = {'drugName': drug, 'condition': word}
        new_rows.append(new_row)

# Create a new DataFrame with the new rows
new_df = pd.DataFrame(new_rows)

# Merge the new DataFrame with the original DataFrame based on 'drugName'
merged_df = pd.merge(new_df, pre_processed_df, on=['drugName', 'condition'], how='outer')

# Display the merged DataFrame
print(merged_df)
```

```python
merged_df.fillna(0, inplace=True)
```
