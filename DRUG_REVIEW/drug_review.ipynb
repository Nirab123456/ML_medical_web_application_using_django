```python
df = pd.read_csv('django.csv')
```

```python
df
```

```python

# Initialize a dictionary to store the word counts
word_counts = {}

# Iterate over the search words
for word in list_of_generic:
    # Count the occurrences of the word in the 'Text' column
    counts = df['generic_name'].str.count(word).sum()
    
    # Store the count in the dictionary
    word_counts[word] = counts

# Print the word counts
```

```python
word_list=[]
for word, count in word_counts.items():
    if count > 0:
        print(f"Word: {word}, Count: {count}")
        word_list.append(word)
```

```python
len(word_list)
```

```python
filtered_df = df_1[df_1['drugName'].isin(word_list)]
new_df = filtered_df.reset_index(drop=True)

```
