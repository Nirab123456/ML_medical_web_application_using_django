```python
import pandas as pd
import numpy as np
import re

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('nirab.csv')

# Convert NaN values to 0
df = df.fillna(0)

# Define a function to calculate the price
def calculate_price(row):
    try:
        match = re.search(r'(\d+\.\d+)', str(row['price']))
        if match:
            price = float(match.group(1))
            if 'x' in str(row['price']):
                match2 = re.search(r'(\d+)\sx\s(\d+)', str(row['price']))
                if match2:
                    left = int(match2.group(1))
                    right = int(match2.group(2))
                    if right != 0:
                        return price / (left * right)
                    else:
                        return 0
            else:
                match2 = re.search(r'(\d+)', str(row['price']))
                if match2:
                    left = int(match2.group(1))
                    if left != 0:
                        return price / left
    except:
        pass
    return 0

# Apply the price calculation function to the 'price' column
df['price'] = df.apply(calculate_price, axis=1)
df['price'] = df['price'].apply(lambda x: round(x, 2))
df = df.applymap(lambda x: str(x).lower())

# Save the changes inplace
df.to_csv('price_fixed.csv', index=False)
```
