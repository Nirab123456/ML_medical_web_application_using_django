```python
import pandas as pd 

df = pd.read_csv('usp_drug_classification.csv')
```

```python
df
```

```python
selected_collumns = ['usp_category', 'usp_class', 'usp_drug','drug_example']
new_df = df[selected_collumns]
```

```python
new_df
```

```python
usp_drug_list = new_df['usp_drug'].unique()
```

```python
len(usp_drug_list)
```

```python
usp_category_list = new_df['usp_category'].unique()
len(usp_category_list)
```

```python
usp_category_list
```

```python
drug_example_list = new_df['drug_example'].unique()
```

```python
django_df = pd.read_csv('django.csv')
```

```python
django_df
```

```python
django_generic_name_list = django_df['generic_name'].unique()
```

```python
#compare 2 list and take the common value

common_drug_list = list(set(drug_example_list) & set(django_generic_name_list))
```

```python
len(common_drug_list)
```

```python

# Step 1: Create a boolean mask for the rows that contain the target words
mask = new_df['drug_example'].isin(common_drug_list)

# Step 2: Filter the DataFrame based on the mask to get related rows
related_rows_df = new_df[mask]

# Step 3: Optionally, reset the index of the new DataFrame
related_rows_df.reset_index(drop=True, inplace=True)

# Display the new DataFrame with related rows
print(related_rows_df)
```
